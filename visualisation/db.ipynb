```python
#from saari.db import *
from fastlite import diagram

from fasthtml.common import database

db = database('data/db/test_fk.db')

class User:
    name:str; 
    pwd:str;

class Paper:
    id: int;
    title: str;
    doi: str;
    abstract: str;
    year: str;
    authors: str;
    publication: str;

class Study:
    id:int;
    user: str;
    title:str;
    lastupdated:int;
    dirty:bool;

class Search:
    id: int;
    query:str;
    study:int;
    title:str;
    lastupdated:int;

class SearchResult:
    search: int;
    paper: int;
    order: int;


users = db.create(User, pk='name')
studies = db.create(Study, pk='id', foreign_keys=['user'])
searches = db.create(Search, pk='id', foreign_keys=['study'])
papers = db.create(Paper, pk='id')
search_results = db.create(SearchResult, pk=['search', 'paper'])

diagram(db.tables)
```
